# 📦 Data Preprocessing — Train/Validation Split

This notebook splits the raw trash image dataset into **training (80%)** and **validation (20%)** sets.

### Pipeline
1. Read images from each class folder in the source directory
2. Randomly shuffle the images to avoid ordering bias
3. Copy **80%** to `data/train/<class>/` and **20%** to `data/val/<class>/`

### 12 Trash Categories
| Category | Category | Category | Category |
|----------|----------|----------|----------|
| 🔋 Battery | 📦 Cardboard | 🔩 Metal | 👟 Shoes |
| 🌿 Biological | 👕 Clothes | 📄 Paper | 🗑️ Trash |
| 🟤 Brown Glass | 🟢 Green Glass | 🧴 Plastic | ⚪ White Glass |

---
## 1. Import Required Libraries

````python
# os       — File and directory path operations
# random   — Shuffling file lists for random train/val split
# shutil   — High-level file copying (preserves metadata)

import os
import random
import shutil
````

---
## 2. Configuration

Define the source directory (raw images organized by class), the destination directories for the split data, and the train/val split ratio.

````python
# Path to the raw dataset folder containing one subfolder per class
# e.g., data set/battery/, data set/plastic/, etc.
source_dir = 'data set'

# Destination directories for the split data
target_dir = "data/train"    # 80% of images go here
val_dir = "data/val"         # 20% of images go here

# Split ratio: 0.8 means 80% training, 20% validation
split_ratio = 0.8
````

---
## 3. Split Dataset into Train and Validation Sets

For each class folder:
1. Create corresponding subdirectories in `data/train/` and `data/val/`
2. Shuffle the files randomly
3. Copy the first 80% to training and the remaining 20% to validation

> **Note:** `shutil.copy2()` is used instead of `shutil.copy()` because it also preserves file metadata (timestamps, permissions).

````python
# Create the top-level train and validation directories
# exist_ok=True prevents errors if the directories already exist
os.makedirs(target_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Iterate through each class directory (e.g., battery, plastic, paper)
for class_name in os.listdir(source_dir):
    class_path = os.path.join(source_dir, class_name)
    
    # Skip any non-directory files in the source folder
    if not os.path.isdir(class_path):
        continue
    
    # Create class-specific subdirectories inside train/ and val/
    # e.g., data/train/battery/ and data/val/battery/
    train_class_dir = os.path.join(target_dir, class_name)
    val_class_dir = os.path.join(val_dir, class_name)
    
    os.makedirs(train_class_dir, exist_ok=True)
    os.makedirs(val_class_dir, exist_ok=True)
    
    # Get all files (images) in the current class directory
    files = os.listdir(class_path)
    
    # Shuffle files randomly so the split is not biased by filename order
    random.shuffle(files)
    
    # Calculate the split index based on the ratio
    # e.g., 1000 images with ratio 0.8 → split_index = 800
    split_index = int(len(files) * split_ratio)
    
    # Split files into training (first 80%) and validation (remaining 20%)
    train_files = files[:split_index]
    val_files = files[split_index:]
    
    # Copy training files to the train directory
    for file in train_files:
        src = os.path.join(class_path, file)
        dst = os.path.join(train_class_dir, file)
        shutil.copy2(src, dst)
    
    # Copy validation files to the validation directory
    for file in val_files:
        src = os.path.join(class_path, file)
        dst = os.path.join(val_class_dir, file)
        shutil.copy2(src, dst)
    
    # Print summary showing the train/val distribution for each class
    print(f"{class_name}: {len(train_files)} train files, {len(val_files)} validation files")
````
