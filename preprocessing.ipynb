```python
import os
import random
import shutil
```

```python
source_dir = 'data set'
target_dir = "data/train"
val_dir = "data/val"

split_ratio = 0.8
```

```python
# Create train and validation directories if they don't exist
os.makedirs(target_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Iterate through each class directory
for class_name in os.listdir(source_dir):
    class_path = os.path.join(source_dir, class_name)
    
    # Skip if it's not a directory
    if not os.path.isdir(class_path):
        continue
    
    # Create class subdirectories in train and validation folders
    train_class_dir = os.path.join(target_dir, class_name)
    val_class_dir = os.path.join(val_dir, class_name)
    
    os.makedirs(train_class_dir, exist_ok=True)
    os.makedirs(val_class_dir, exist_ok=True)
    
    # Get all files in the class directory
    files = os.listdir(class_path)
    
    # Shuffle the files for random split
    random.shuffle(files)
    
    # Calculate split index
    split_index = int(len(files) * split_ratio)
    
    # Split files into train and validation
    train_files = files[:split_index]
    val_files = files[split_index:]
    
    # Copy training files
    for file in train_files:
        src = os.path.join(class_path, file)
        dst = os.path.join(train_class_dir, file)
        shutil.copy2(src, dst)
    
    # Copy validation files
    for file in val_files:
        src = os.path.join(class_path, file)
        dst = os.path.join(val_class_dir, file)
        shutil.copy2(src, dst)
    
    print(f"{class_name}: {len(train_files)} train files, {len(val_files)} validation files")
```
